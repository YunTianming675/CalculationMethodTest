```python
import numpy
import matplotlib.pyplot as plt
import sympy
```

```python
x1 = numpy.linspace(10, 100, 10, dtype=int)
y1 = [62, 68, 75, 81, 89, 95, 102, 108, 115, 122]
```

```python
plt.scatter(x1, y1)
```

```python
# 最小二乘法拟合
coeff1 = numpy.array([[len(x1), numpy.sum(x1)], [numpy.sum(x1), numpy.sum(x1 ** 2)]])
coeff2 = numpy.array([[numpy.sum(y1)], [numpy.sum(x1 * y1)]])
m1 = numpy.matrix(coeff1)
m2 = numpy.matrix(coeff2)
print(m1)
print(m2)
```

```python
m3 = m1.I * m2
a3 = numpy.array(m3)
print(m3)
```

```python
x = sympy.symbols("x")
f1 = a3[0][0] * x + a3[1][0]
print("拟合直线：", f1)
```

```python
x2 = numpy.linspace(10, 100, 901)
y2 = list()
for _, i in enumerate(x2):
    y2.append(f1.evalf(subs={x:i}))
print(y2)
```
